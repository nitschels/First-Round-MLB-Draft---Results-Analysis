# Defining a Flop

This is where my definitions WILL get a little bit testy. I need to make some assumptions and definitions of what I deem as a '*flop*' for a first round draft pick. Firstly, any player drafted in the 2020 draft or later, will not be identified a flop, since players drafted out of high school have only played in the minors for at most 5 years, which puts them around age 23 at most in 2026, and the average debut age is between 23 and 25 for such players. Now, I will preface this by saying that just because a player is a flop, does not mean that they didn't have a good MLB career, it's just that they didn't amount to what they could have based on what is expected of a first round talent.

If a player hasn't made their MLB debut, while being drafted pre-2020, I have deemed them as a flop. Even with the 2019 draft class, the impact these players could have made at this point puts them behind the curve already, and otherwise 25+ year olds in the minor leagues still will typically stay there. Therefore, I will also identify those not having made their debut at 25 (typically older than the average MLB debut) as a flop. This is where it gets a little bit more difficult to identify, and some may argue this. If the player has below a 0.8 WAR, then he is a career bench player (on average), worse than is expected of a first round pick. Also, if the player has a short career (below 5 years), he is also a 'flop'.

The 'flop' variable will be used to see how successful a draft class is, in its entirety.

```python
# if years in minors NaN -> flop
# else:
  # if high school:
    # if years in minors > 6 and years in majors < 5
      # flop
    #elif WAR by Season < .8
      # flop
    # else:
      # no flop
  # else (college player):
    # if years in minors > 5 and years in majors < 5
      # flop
    # elif WAR by Season < .8
      # flop
    # else:
      # no flop
combined_df_copy = combined_df
combined_df['Flop?'] = np.nan




for index, player in combined_df.iterrows():
  if player['Year'] >= 2020:
    combined_df.at[index, 'Flop?'] = 0
  elif pd.isna(player['Most Recent Season']):
    combined_df.at[index, 'Flop?'] = 1
  else:
    if player['HS?'] == 1:
      if player['WAR by Season'] < 0.2:
        combined_df.at[index, 'Flop?'] = 1

      elif player['Years in Minors'] <= 7 and player['Total Seasons'] >= 5:
        combined_df.at[index, 'Flop?'] = 0
      else:
        combined_df.at[index, 'Flop?'] = 1
    else:
      if player['WAR by Season'] < 0.3:
        combined_df.at[index, 'Flop?'] = 1
      elif player['Years in Minors'] <= 5 and player['Total Seasons'] >= 5:
        combined_df.at[index, 'Flop?'] = 0
      else:
        combined_df.at[index, 'Flop?'] = 1
combined_df.head(20)
```

# Highest Levels Reached

```python
level_mapping = {
    'A' : 1,
    'A-' : 1.33,
    'A+' : 1.66,
    'AA': 2,
    'AAA' : 3,
    'MLB' : 4
}

combined_df['Numerical Level'] = combined_df['High Level'].map(level_mapping)
```

Here we have data from the past 25 drafts, looking at the average level by year. As we can see, there is a dropp off in the average level in the last few years of the draft, however, this can be attributed to the players still being young and needing time to develop

```python
yearly_level = combined_df.groupby(combined_df['Year'])['Numerical Level'].mean()


plt.plot(yearly_level.index, yearly_level.values)


positions = [1,2,3,4]
labels = ['A', 'AA', 'AAA', 'MLB']

plt.title("Average level by draft year")
plt.ylim(0,4)
plt.ylabel("Average Level")
plt.xlabel("Year")
plt.yticks(positions, labels)
plt.show()
```

If we look at the years leading up to the 2020 draft, we see a different trend

```python
yearly_level_pre2021 = yearly_level[yearly_level.index < 2021]
yearly_level_pre2021

x = yearly_level_pre2021.index.to_numpy()
y = yearly_level_pre2021.values

m,b = np.polyfit(x,y, 1)

correlation_matrix = np.corrcoef(x, y)
r_value = correlation_matrix[0,1]
r_2 = r_value ** 2

plt.scatter(yearly_level_pre2021.index, yearly_level_pre2021.values)


plt.plot(x, m*x + b, color = 'red')
plt.ylim(0,4)

plt.ylabel("Average Level")
plt.xlabel("Year")
plt.yticks(positions, labels)

text_str = f'Line: y = {m:.4f}x + {b:.2f}\nCorrelation (r): {r_value:.3f}\nR²: {r_2:.3f}'
plt.gca().text(0.55, 0.75, text_str, transform=plt.gca().transAxes,
            fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
plt.legend(loc = 'lower right')

plt.show()
```

```python
hs_df = combined_df[combined_df['HS?'] == 1]
hs_by_year = hs_df.groupby('Year').agg(
    total_hs_players = ('HS?', 'count'),
    avg_hs_debut_year = ('Debut Year', 'mean'),
    hs_made_majors= ('Debut Year', lambda x: x.notna().sum())
)

hs_by_year['avg_hs_years_minors'] = hs_by_year['avg_hs_debut_year'] - hs_by_year.index
hs_by_year['hs_pct_majors'] = hs_by_year['hs_made_majors'] / hs_by_year['total_hs_players']
hs_by_year
```

```python
col_df = combined_df[combined_df['HS?'] == 0]
col_by_year = col_df.groupby('Year').agg(
    total_college_players = ('HS?', 'count'),
    college_debut_year = ('Debut Year', 'mean'),
    college_made_majors= ('Debut Year', lambda x: x.notna().sum())
)

col_by_year['college_years_in_minors'] = col_by_year['college_debut_year'] - col_by_year.index
col_by_year['pct_college_majors'] = col_by_year['college_made_majors'] / col_by_year['total_college_players']
col_by_year
```

```python
merged_df = pd.merge(hs_by_year, col_by_year, on = 'Year')
merged_df['pct_college_majors'] = merged_df['pct_college_majors'] * 100
merged_df['hs_pct_majors'] = merged_df['hs_pct_majors'] * 100
```

```python
def get_flops(df: pd.DataFrame):
  result_df = df.groupby("Year").agg(
      flop = ('Flop?', 'sum'),
      total_drafted = ('Rnd', 'count')
  ).assign(flop_rate = lambda x: x['flop'] / x['total_drafted'])
  result_df['flop_rate'] = result_df['flop_rate'] * 100
  return result_df
```

```python
coll_pitchers = pitchers_df[pitchers_df['HS?'] == 0]
coll_pitchers_df = get_flops(coll_pitchers)
coll_pitchers_df = coll_pitchers_df.rename(columns = {'flop': 'college_pitcher_flops', 'total_drafted': 'college_pitchers_drafted', 'flop_rate' : 'college_pitcher_flop_rate'})

hs_pitchers = pitchers_df[pitchers_df['HS?'] == 1]
hs_pitchers_df = get_flops(hs_pitchers)
hs_pitchers_df = hs_pitchers_df.rename(columns = {'flop': 'hs_pitcher_flops', 'total_drafted': 'hs_pitchers_drafted', 'flop_rate' : 'hs_pitcher_flop_rate'})

coll_hitters = hitters_df[hitters_df['HS?'] == 0]
coll_hitters_df = get_flops(coll_hitters)
coll_hitters_df = coll_hitters_df.rename(columns = {'flop': 'college_hitter_flops', 'total_drafted': 'college_hitters_drafted', 'flop_rate' : 'college_hitter_flop_rate'})

hs_hitters = hitters_df[hitters_df['HS?'] == 1]
hs_hitters_df = get_flops(hs_hitters)
hs_hitters_df = hs_hitters_df.rename(columns = {'flop': 'hs_hitter_flops', 'total_drafted': 'hs_hitters_drafted', 'flop_rate' : 'hs_hitter_flop_rate'})

combined_flop_df = pd.concat([coll_pitchers_df, hs_pitchers_df, coll_hitters_df, hs_hitters_df], axis = 1)
combined_flop_df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
combined_flop_df = combined_flop_df[combined_flop_df.index < 2020]

x = combined_flop_df.index.to_numpy()

plot_configs = [
    (axs[0, 0], 'college_pitcher_flop_rate', 'College Pitcher Flop Rate', 'royalblue', 'o', 'College Pitchers'),
    (axs[1, 0], 'hs_pitcher_flop_rate',      'HS Pitcher Flop Rate',      'orange',    's','High School Pitchers'),
    (axs[0, 1], 'college_hitter_flop_rate',  'College Hitter Flop Rate',  'red',       'D', 'College Hitters'),
    (axs[1, 1], 'hs_hitter_flop_rate',       'HS Hitter Flop Rate',       'gray',      '^', 'High School Hitters'),
]

for ax, col, label, color, marker, title in plot_configs:
    y = combined_flop_df[col]

    m, b = np.polyfit(x, y, 1)

    r_value = np.corrcoef(x, y)[0, 1]
    r_2 = r_value ** 2

    ax.plot(x, m * x + b, color='red')
    ax.scatter(combined_flop_df.index, y, label=label, color=color, marker=marker)
    ax.legend()
    ax.set_title(title)
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))

    ax.text(
        0.05, 0.95,
        f'$R^2$ = {r_2:.3f}\nSlope = {m:.3f}',
        transform=ax.transAxes,
        verticalalignment='top',
        fontsize=10,
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7)
    )

fig.supxlabel('Year')
fig.supylabel('Players that are Flops (%)')
fig.suptitle('Flop Rate by Position and Draft Level')

plt.setp(axs, ylim=(0, 105))
plt.tight_layout()
plt.show()
```
